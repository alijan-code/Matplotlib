```python
Scatter plot
```

```python
x = [1,2,3,4,5,6]
y = [4,2,3,5,1,8]
plt.title("Scatter plot")
plt.xlabel("month")
plt.ylabel("numbers")
plt.scatter(x,y)
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [4,2,3,5,1,8]
plt.title("Scatter plot")
plt.xlabel("month")
plt.ylabel("numbers")

c = ['red', 'magenta', 'green', 'orange', 'blue', 'brown']

plt.scatter(x,y, color = c, s = 150)
plt.show()
```
